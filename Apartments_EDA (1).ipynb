# import libraries

```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.options.display.float_format = '{:,.2f}'.format  # set other global format
```

```python
df=pd.read_csv("/content/apartments_updated.csv")
```

```python
df.head(2)
```

```python
df.columns=df.columns.str.lower()
df.columns
```

## Drop unwanted columns

```python
df.drop(['url','details','number_of_pages'],axis=1,inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop_duplicates(subset='link',inplace=True)
```

```python
df.info()
```

## Drop NA Values in dataset

```python
pd.DataFrame.dropna(df,axis=0,subset='pathrooms',inplace=True)
```

```python
pd.DataFrame.dropna(df,axis=0,subset='address',inplace=True)
```

```python
df.info()
```

# dropped string values in bedrooms and pathrooms

```python
df.drop(index=df[df['bedrooms']=='label.rooms.plus10'].index,inplace=True)
df.drop(index=df[df['pathrooms']=='label.baths.plus6'].index,inplace=True)
```

```python
df['pathrooms']=df.pathrooms.astype(int)
df['bedrooms']=df['bedrooms'].astype(int)
```

```python
df.info()
```

```python
df.drop(index=df[df['price']=='اتصل لمعرفة السعر'].index,inplace=True)
```

```python
df['price']=df['price'].str.replace('جنيه','').str.replace(",","").astype('Int64')
```

```python
df.price.info()
```

```python
df['area']=df['area'].str.replace("متر²","").astype(float)
```

```python
df.info()
```

## Droping for area biger than 700 and bedrooms > 8 and pathrooms > 8

```python
df.drop(index=df[df['area']>700].index,inplace=True)
df.drop(index=df[df['pathrooms']>8].index,inplace=True)
df.drop(index=df[df['bedrooms']>8].index,inplace=True)
df.drop(index=df[df['price']<100000].index,inplace=True)
df.drop(index=df[df['area']<50].index,inplace=True)
```
