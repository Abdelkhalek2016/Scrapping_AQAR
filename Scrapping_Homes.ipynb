```python
import requests
from bs4 import BeautifulSoup

url = "https://aqarmap.com.eg/ar/for-sale/apartment/cairo/"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    listings = soup.find_all("div", class_="sectionBody ng-scope")
    
    for listing in listings:
        title = listing.find("h2", class_="listing-item-title").text.strip()
        price = listing.find("span", class_="listing-price").text.strip()
        location = listing.find("span", class_="listing-location").text.strip()
        
        print("Title:", title)
        print("Price:", price)
        print("Location:", location)
        print("-" * 30)
else:
    print("Failed to fetch the page")
```
